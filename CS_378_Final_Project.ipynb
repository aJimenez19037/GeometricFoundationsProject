```python
# Actor network
class Actor(nn.Module):
    def __init__(self, state_dim, action_dim, hidden_dim=256):
        super(Actor, self).__init__()
        self.fc1 = nn.Linear(state_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        self.out = nn.Linear(hidden_dim, action_dim)

    def forward(self, state):
        x = torch.relu(self.fc1(state))
        x = torch.relu(self.fc2(x))
        logits = self.out(x)
        probs = torch.softmax(logits, dim=-1)  # Discrete action probabilities
        return probs

# Critic network (Q-value function)
class Critic(nn.Module):
    def __init__(self, state_dim, action_dim, hidden_dim=256):
        super(Critic, self).__init__()
        self.fc1 = nn.Linear(state_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        self.out = nn.Linear(hidden_dim, action_dim)  # Outputs Q-value for each action

    def forward(self, state):
        x = torch.relu(self.fc1(state))
        x = torch.relu(self.fc2(x))
        q_values = self.out(x)
        return q_values
```

```python
class SACAgent:
    def __init__(self, state_dim, action_dim, device):
        self.device = device

        self.actor = Actor(state_dim, action_dim).to(device)
        self.critic1 = Critic(state_dim, action_dim).to(device)
        self.critic2 = Critic(state_dim, action_dim).to(device)
        self.target_critic1 = Critic(state_dim, action_dim).to(device)
        self.target_critic2 = Critic(state_dim, action_dim).to(device)

        self.actor_optimizer = optim.Adam(self.actor.parameters(), lr=3e-4)
        self.critic1_optimizer = optim.Adam(self.critic1.parameters(), lr=3e-4)
        self.critic2_optimizer = optim.Adam(self.critic2.parameters(), lr=3e-4)

        # Copy weights to target networks
        self.target_critic1.load_state_dict(self.critic1.state_dict())
        self.target_critic2.load_state_dict(self.critic2.state_dict())

        self.gamma = 0.99
        self.tau = 0.005
        self.alpha = 0.2  # Entropy regularization coefficient

        self.memory = deque(maxlen=100000)
        self.batch_size = 64

    def select_action(self, state, eval=False):
        state = torch.FloatTensor(state).unsqueeze(0).to(self.device)
        probs = self.actor(state).cpu().detach().numpy().flatten()
        if eval:
            action = np.argmax(probs)
        else:
            action = np.random.choice(len(probs), p=probs)
        return action

    def store(self, transition):
        self.memory.append(transition)

    def sample(self):
        batch = random.sample(self.memory, self.batch_size)
        state, action, reward, next_state, done = map(np.array, zip(*batch))
        return (
            torch.FloatTensor(state).to(self.device),
            torch.LongTensor(action).to(self.device),
            torch.FloatTensor(reward).unsqueeze(1).to(self.device),
            torch.FloatTensor(next_state).to(self.device),
            torch.FloatTensor(done).unsqueeze(1).to(self.device)
        )

    def update(self):
        if len(self.memory) < self.batch_size:
            return

        state, action, reward, next_state, done = self.sample()

        # Critic loss
        with torch.no_grad():
            next_action_probs = self.actor(next_state)
            next_q1 = self.target_critic1(next_state)
            next_q2 = self.target_critic2(next_state)
            next_q = torch.min(next_q1, next_q2)
            next_value = (next_action_probs * (next_q - self.alpha * torch.log(next_action_probs + 1e-10))).sum(dim=1, keepdim=True)
            target_q = reward + (1 - done) * self.gamma * next_value

        # Current Q estimates
        current_q1 = self.critic1(state).gather(1, action.unsqueeze(1))
        current_q2 = self.critic2(state).gather(1, action.unsqueeze(1))

        critic1_loss = nn.MSELoss()(current_q1, target_q)
        critic2_loss = nn.MSELoss()(current_q2, target_q)

        self.critic1_optimizer.zero_grad()
        critic1_loss.backward()
        self.critic1_optimizer.step()

        self.critic2_optimizer.zero_grad()
        critic2_loss.backward()
        self.critic2_optimizer.step()

        # Actor loss
        probs = self.actor(state)
        log_probs = torch.log(probs + 1e-10)
        q1 = self.critic1(state)
        q2 = self.critic2(state)
        q = torch.min(q1, q2)
        actor_loss = (probs * (self.alpha * log_probs - q)).sum(dim=1).mean()

        self.actor_optimizer.zero_grad()
        actor_loss.backward()
        self.actor_optimizer.step()

        # Update target networks
        for target_param, param in zip(self.target_critic1.parameters(), self.critic1.parameters()):
            target_param.data.copy_(self.tau * param.data + (1 - self.tau) * target_param.data)
        for target_param, param in zip(self.target_critic2.parameters(), self.critic2.parameters()):
            target_param.data.copy_(self.tau * param.data + (1 - self.tau) * target_param.data)
```

```python
def euclidean(a, b):
    return math.hypot(a[0] - b[0], a[1] - b[1])

class MotionPredictiveSafetyController:
    """
    Motion Predictive Safety Controller for discrete-action SAC.
    Predicts trajectories of ego and surrounding vehicles and replaces
    unsafe actions with safe alternatives.
    """
    def __init__(self, env, horizon=7, dt=None, min_distance=2.0):
        self.env = env
        # Prediction horizon (number of policy steps)
        self.horizon = horizon
        # Timestep between policy actions
        self.dt = dt or 1.0 / env.unwrapped.config.get("policy_frequency", 5)
        # Minimum safety distance (meters)
        self.min_distance = min_distance
        # Observation shape (V vehicles, F features)
        self.obs_shape = env.observation_space.shape
        # IDM parameters for other vehicles
        self.a_max = 1.0    # max acceleration
        self.b = 1.5        # comfortable braking
        self.v0 = 25.0      # desired speed
        self.delta = 4      # acceleration exponent
        self.T_head = 1.5   # desired time headway
        self.s0 = 2.0       # minimum gap offset
        self.eps = 1e-3     # small value to avoid divide-by-zero

    def idm_acc(self, v, v_lead, gap):
        # Clamp inputs to avoid overflow/nan
        gap = max(gap, self.s0 + self.eps)
        v = max(v, 0.0)
        v_lead = max(v_lead, 0.0)
        v_ratio = min(v / self.v0, 1.0)
        # Desired dynamic gap
        s_star = self.s0 + v * self.T_head + max(0.0,
            (v * (v - v_lead)) / (2.0 * math.sqrt(self.a_max * self.b) + self.eps))
        # Safe power and division
        term1 = 1.0 - v_ratio**self.delta
        term2 = (s_star / gap)**2
        acc = self.a_max * (term1 - term2)
        # Clip acceleration to physical bounds
        return float(np.clip(acc, -self.b, self.a_max))

    def _unflatten(self, state):
        # Reshape flat state vector into VxF matrix
        return state.reshape(self.obs_shape)

    def predict_other_trajectories(self, obs_matrix):
        # Predict other vehicles with IDM + constant lane
        V, F = obs_matrix.shape
        others = obs_matrix[1:]
        trajs = {i+1: [] for i in range(V-1)}
        sim = others.copy()
        for _ in range(self.horizon):
            new_sim = []
            for idx, (x, y, vx, vy) in enumerate(sim):
                # approximate gap to vehicle ahead (use s0 if none)
                gap = abs(y - 1.0)
                a_lon = self.idm_acc(vy, vy, gap)
                vy_new = vy + a_lon * self.dt
                # Prevent non-finite speeds
                vy_new = float(np.nan_to_num(vy_new, nan=0.0, posinf=self.v0, neginf=0.0))
                y_new = y + vy_new * self.dt
                x_new = x
                new_sim.append([x_new, y_new, vx, vy_new])
                trajs[idx+1].append((x_new, y_new))
            sim = np.array(new_sim)
        return trajs

    def predict_ego_trajectory(self, obs_matrix, action):
        # Discrete actions: 0=left,1=idle,2=right,3=faster,4=slower
        ego = obs_matrix[0].copy()
        x, y, vx, vy = ego
        lane_width = self.env.unwrapped.config.get("lane_width", 4.0)
        acc_cmd = 0.0
        lat_cmd = 0.0
        accel = 2.0
        if action == 3:
            acc_cmd = accel
        elif action == 4:
            acc_cmd = -accel
        elif action == 0:
            lat_cmd = -lane_width / self.horizon
        elif action == 2:
            lat_cmd = lane_width / self.horizon
        traj = []
        for _ in range(self.horizon):
            vy = max(vy + acc_cmd * self.dt, 0.0)
            y += vy * self.dt
            x += lat_cmd
            traj.append((x, y))
        return traj

    def is_safe(self, state, action):
        obs = self._unflatten(state)
        ego_traj = self.predict_ego_trajectory(obs, action)
        other_traj = self.predict_other_trajectories(obs)
        for t in range(self.horizon):
            ex, ey = ego_traj[t]
            for traj in other_traj.values():
                ox, oy = traj[t]
                if euclidean((ex, ey), (ox, oy)) < self.min_distance:
                    return False
        return True

    def get_safe_action(self, state):
        """
        Return the action that maximizes minimum distance if safe,
        otherwise idle (1).
        """
        best_action, best_margin = 1, -float('inf')
        obs = self._unflatten(state)
        for a in range(self.env.action_space.n):
            if not self.is_safe(state, a):
                continue
            ego_traj = self.predict_ego_trajectory(obs, a)
            other_traj = self.predict_other_trajectories(obs)
            margin = min(
                euclidean(ego_traj[t], traj[t])
                for t in range(self.horizon)
                for traj in other_traj.values()
            )
            if margin > best_margin:
                best_margin, best_action = margin, a
        return best_action
```

```python
def add_sensor_noise(obs, noise_std=0.5):
    # TODO make more complex sensor noise model
    # Add Gaussian noise to all features
    return obs + np.random.normal(0, noise_std, size=obs.shape)
```

```python
def make_roundabout_env(mode: str, obs_count: int = 8) -> gym.Env:
    spawn_counts = {
        'easy':   8,
        'medium': 12,
        'hard':   20,
    }
    if mode not in spawn_counts:
        raise ValueError(f"Unknown mode {mode!r}, expected one of {list(spawn_counts)}")

    cfg = {
        # vehicles to actually spawn into the road:
        "vehicles_count": spawn_counts[mode],

        # What the agent sees — keep fixed so neural nets keep the same input dimension
        "observation": {
            "type":           "Kinematics",
            "vehicles_count": obs_count,
            "features":       ["x", "y", "vx", "vy"],
        },

        "action": {
            "type": "DiscreteMetaAction",
        },
        "duration": 40,
        "other_vehicles_type": "highway_env.vehicle.behavior.IDMVehicle"
    }

    return gym.make("roundabout-v0", render_mode="rgb_array", config=cfg)

```

```python
def collect_nf_data(env, num_steps):
    noisy, true = [], []
    obs, _ = env.reset()
    for _ in range(num_steps):
        true_obs = obs.copy()
        noisy_obs = add_sensor_noise(obs)
        noisy.append(noisy_obs.flatten())
        true.append(true_obs.flatten())
        action = env.action_space.sample()  # or your SAC policy
        obs, _, done, _, _ = env.step(action)
        if done:
            obs, _ = env.reset()
    return np.array(noisy), np.array(true)
```

```python
def create_nf_model(input_dim):
    transforms = []
    for i in range(5):
        # Alternating binary mask
        mask = torch.tensor([(j + i) % 2 for j in range(input_dim)], dtype=torch.bool)

        def net_fn(in_features, out_features):
            return ResidualNet(
                in_features=in_features,
                out_features=out_features,
                hidden_features=64,
                context_features=input_dim,  #  net to expect conditioning
                num_blocks=2
            )

        transforms.append(AffineCouplingTransform(
            mask=mask,
            transform_net_create_fn=net_fn
        ))
        transforms.append(ReversePermutation(features=input_dim))

    transform = CompositeTransform(transforms)
    base_dist = StandardNormal(shape=[input_dim])
    return Flow(transform, base_dist)
```

Train agent with MSPC and NF

```python
def correct_observation(nf_model, noisy_obs, scaler_x, scaler_y, device):
    # flatten & scale
    x_raw = noisy_obs.flatten()[None, :]
    x_scaled = scaler_x.transform(x_raw)

    x_t = torch.tensor(x_scaled, dtype=torch.float32, device=device)

    # get the MAP: push z=0 through the inverse
    nf_model.eval()
    with torch.no_grad():
        z0 = torch.zeros(1, x_t.size(1), device=device)
        y_scaled_map, _ = nf_model._transform.inverse(z0, context=x_t)

    # inverse‐scale
    y_scaled = y_scaled_map.cpu().numpy()
    y_map    = scaler_y.inverse_transform(y_scaled)

    return y_map.squeeze(0)
```

```python
scaler_x = StandardScaler().fit(noisy_all)
scaler_y = StandardScaler().fit( true_all)

flow = create_nf_model(state_dim).to(device)
flow.load_state_dict(torch.load("flow.pt", map_location=device))
flow.eval()   # switch to eval mode since we’re sampling/denoising

episodes_per_mode = 200
obs_count = 8   # dont change
env = make_roundabout_env("easy", obs_count=obs_count)
episodes_without_mpsc = episodes_per_mode // 2
# Initialize once, before curriculum
state_dim = env.observation_space.shape[0] * env.observation_space.shape[1]
action_dim = env.action_space.n
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
agent = SACAgent(state_dim, action_dim, device)

tn = 7  # prediction horizon for MPSC

for mode in ['easy', 'medium', 'hard']:
    print(f"\n=== Training on {mode.upper()} ({episodes_per_mode} eps) ===")
    env  = make_roundabout_env(mode)
    dt   = 1.0 / env.unwrapped.config["policy_frequency"]
    mpsc = MotionPredictiveSafetyController(env, horizon=tn, dt=dt)

    noisy_obs_list, true_obs_list = [], []
    safe_count, total_count = 0, 0

    for ep in range(episodes_per_mode):
        obs, _         = env.reset()
        episode_reward = 0.0
        done           = False

        while not done:
            # Get true & noisy obs
            true_obs  = obs.copy()
            noisy_obs = add_sensor_noise(obs)

            # Denoise before decision
            denoised   = correct_observation(flow, noisy_obs,scaler_x,scaler_y, device)
            state      = denoised.flatten()

            # Pick action
            a = agent.select_action(state)
            if ep < episodes_without_mpsc:
                safe_a = a
            else:
                if mpsc.is_safe(state, a):
                    safe_count += 1
                    safe_a = a
                else:
                    safe_a     = mpsc.get_safe_action(state)
                total_count += 1

            # Step environment
            next_obs, reward, term, trunc, _ = env.step(safe_a)
            next_noisy_obs = add_sensor_noise(next_obs)

            done = term or trunc

            # Denoise next obs
            next_denoised = correct_observation(flow, next_noisy_obs, scaler_x, scaler_y, device)
            next_state    = next_denoised.flatten()

            # Store & update SAC
            agent.store((state, safe_a, reward, next_state, float(done)))
            agent.update()

            # Log NF data
            noisy_obs_list.append(noisy_obs.flatten())
            true_obs_list.append(true_obs.flatten())

            obs = next_obs
            episode_reward += reward

        print(f"[{mode[:1].upper()}] Ep {ep+1:3d}/{episodes_per_mode}  Reward: {episode_reward:.2f}")

    pct = safe_count / max(total_count,1) * 100
    print(f"[{mode[:1].upper()}] Safety interventions: {pct:.1f}%")

    # checkpoint agent
    # torch.save(agent.state_dict(), f"sac_roundabout_complex_{mode}.pt")
    with open(f"sac_agent_complex_{mode}.pkl", "wb") as f:
          pickle.dump(agent, f)
    # save NF training data
    with open(f"nf_data_complex_{mode}.pkl", "wb") as f:
        pickle.dump((noisy_obs_list, true_obs_list), f)
```

```python
# Not working
class DoneWrapper(gym.Wrapper):
    def step(self, action):
        out = self.env.step(action)
        if len(out) == 5:
            obs, reward, term, trunc, info = out
            return obs, reward, term or trunc, info
        return out

    def reset(self, **kwargs):
        out = self.env.reset(**kwargs)
        if isinstance(out, tuple) and len(out) == 2:
            obs, info = out
            return obs, info
        return out, {}

class RewardNormWrapper(gym.RewardWrapper):
    def __init__(self, env, low=-1.0, high=1.0):
        super().__init__(env)
        self.low, self.high = low, high
    def reward(self, reward):
        r = (reward - self.low) / (self.high - self.low)
        return float(np.clip(r, 0.0, 1.0))

def make_roundabout_env(mode, obs_count=8):
    spawn = {'easy':8,'medium':12,'hard':20}
    cfg = {
        "vehicles_count": spawn[mode],
        "observation": {
            "type":"Kinematics",
            "vehicles_count":obs_count,
            "features":["x","y","vx","vy"],
        },
        "action":{"type":"DiscreteMetaAction"},
        "duration":40,
        "other_vehicles_type":"highway_env.vehicle.behavior.IDMVehicle",
    }
    base = gym.make("roundabout-v0", render_mode="rgb_array", config=cfg)
    return DoneWrapper(RewardNormWrapper(base, low=-1.0, high=1.0))

# --- evaluation with a smaller planning budget ---
def evaluate(agent_cls, modes, n_episodes=30, obs_count=8, budget=20):
    stats = {}
    for mode in modes:
        env = make_roundabout_env(mode, obs_count)

        # Pass a smaller budget into the agent’s config:
        cfg = {
            "models": [[]],        # one model: the true env
            "budget": budget,      # ← throttle here!
        }
        agent = agent_cls(env, config=cfg)

        returns = []
        for _ in range(n_episodes):
            agent.reset()
            obs, _ = env.reset()
            done = False
            R = 0.0
            while not done:
                action, *_ = agent.plan(obs)
                obs, reward, done, _ = env.step(action)
                R += reward
            returns.append(R)

        stats[mode] = {
            "mean": np.mean(returns),
            "std":  np.std(returns),
            "min":  np.min(returns),
            "max":  np.max(returns),
        }
    return stats

modes = ["easy","medium","hard"]
stats = evaluate(DiscreteRobustPlannerAgent, modes, n_episodes=5, budget=20)
for mode in modes:
    s = stats[mode]
    print(f"{mode.capitalize():>6} | mean={s['mean']:.2f}, "
          f"min={s['min']:.2f}, max={s['max']:.2f}, std={s['std']:.2f}")
```

```python
# 4‑tuple wrapper
class DoneWrapper(gym.Wrapper):
    def step(self, action):
        obs, reward, term, trunc, info = self.env.step(action)
        return obs, reward, term or trunc, info
    def reset(self, **kwargs):
        return self.env.reset(**kwargs)

# Reward normalizer
class RewardNormWrapper(gym.RewardWrapper):
    def __init__(self, env, low=-1.0, high=1.0):
        super().__init__(env)
        self.low, self.high = low, high
    def reward(self, r):
        return float(np.clip((r - self.low)/(self.high - self.low), 0, 1))

def make_roundabout_env(mode="hard", obs_count=8):
    spawn = {'easy':8,'medium':12,'hard':20}
    cfg = {
        "vehicles_count": spawn[mode],
        "observation": {"type":"Kinematics","vehicles_count":obs_count,"features":["x","y","vx","vy"]},
        "action": {"type":"DiscreteMetaAction"},
        "duration": 40,
        "other_vehicles_type":"highway_env.vehicle.behavior.IDMVehicle",
    }
    base = gym.make("roundabout-v0", render_mode="rgb_array", config=cfg)
    return DoneWrapper(RewardNormWrapper(base))

agent_files = {
    "sac_agent_simple_true_hard.pkl":  "SAC (true obs)",
    "sac_agent_normal_true_hard.pkl":  "SAC+MPSC (true obs)",
    "sac_agent_simple_hard.pkl":       "SAC (noisy obs)",
    "sac_agent_normal_hard.pkl":       "SAC+MPSC (noisy obs)",
    "sac_agent_complex_hard.pkl":      "SAC+MPSC+NF (noisy obs)"
}

def run_episode(agent, env):
    # get horizon
    T_max = env.unwrapped.config["duration"]
    obs, _ = env.reset()
    traj = np.zeros((T_max, 2))
    rewards = np.zeros(T_max)
    actions = np.zeros(T_max, dtype=int)

    done = False
    last_pos = (obs.flatten()[0], obs.flatten()[1])

    for t in range(T_max):
        if not done:
            flat = obs.flatten()
            x,y = flat[0], flat[1]
            last_pos = (x,y)
            a = agent.select_action(flat, eval=True)
            obs, r, done, _ = env.step(a)
        else:
            r = 0
            a = 0

        traj[t] = last_pos
        rewards[t] = r
        actions[t] = a

    return traj, rewards, actions

# Evaluate N episodes per agent in the hard environment
N = 30
mode = "hard"
results = {}
for path, label in agent_files.items():
    env = make_roundabout_env(mode)
    with open(path, "rb") as f:
        agent = pickle.load(f)
    if hasattr(agent, "config"):
        agent.config["models"] = [[]]

    all_trajs = np.zeros((N, env.unwrapped.config["duration"], 2))
    all_rews  = np.zeros((N, env.unwrapped.config["duration"]))
    all_acts  = []

    for i in range(N):
        traj, rews, acts = run_episode(agent, env)
        all_trajs[i] = traj
        all_rews[i]  = rews
        all_acts.append(acts)

    all_acts = np.concatenate(all_acts)
    results[label] = {"traj": all_trajs, "rew": all_rews, "act": all_acts}

plt.figure(figsize=(8,6))
for label, data in results.items():
    xs = data["traj"][...,0]
    ys = data["traj"][...,1]
    mean_x = xs.mean(0); mean_y = ys.mean(0)
    min_y, max_y = ys.min(0), ys.max(0)
    plt.plot(mean_x, mean_y, label=label)
    plt.fill_between(mean_x, min_y, max_y, alpha=0.1)
plt.title("Mean Ego Trajectory ± range over 30 eps")
plt.xlabel("X"); plt.ylabel("Y")
plt.legend(); plt.grid(True)
plt.show()

plt.figure(figsize=(8,4))
steps = np.arange(env.unwrapped.config["duration"])
for label, data in results.items():
    m = data["rew"].mean(0)
    s = data["rew"].std(0)
    plt.plot(steps, m, label=label)
    plt.fill_between(steps, m-s, m+s, alpha=0.1)
plt.title("Mean Reward per Step ±1 std")
plt.xlabel("Step"); plt.ylabel("Normalized Reward")
plt.legend(); plt.grid(True)
plt.show()

plt.figure(figsize=(8,4))
bins = np.arange(env.action_space.n+1)-0.5
for label, data in results.items():
    plt.hist(data["act"], bins=bins, histtype="step", label=label)
plt.title("Action Distribution over 30 eps")
plt.xlabel("Action"); plt.ylabel("Count")
plt.legend(); plt.grid(True)
plt.show()
```

```python

def record_videos(env, video_folder, agent=None, fps=15):
    """Wrap `env` so every render() is recorded, then run exactly one episode
    (agent.select_action(obs.flatten()) if agent provided, else random) and
    return the path to the saved MP4."""
    os.makedirs(video_folder, exist_ok=True)
    recorder = RecordVideo(env, video_folder=video_folder, episode_trigger=lambda e: True)
    recorder.unwrapped.set_record_video_wrapper(recorder)

    # reset env
    out = recorder.reset()
    obs = out[0] if isinstance(out, tuple) else out
    done = False

    # rollout loop
    while not done:
        if agent is None:
            action = recorder.action_space.sample()
        else:
            action = agent.select_action(obs.flatten(), eval=True)
        step = recorder.step(action)
        # unpack Gym vs Gymnasium
        if len(step) == 5:
            obs, _, term, trunc, _ = step
            done = term or trunc
        else:
            obs, _, done, _ = step
        recorder.render()  # **this** export intermediate 15 Hz frames

    recorder.close()

    # grab the latest MP4
    mp4s = sorted(glob.glob(os.path.join(video_folder, "*.mp4")))
    if not mp4s:
        raise FileNotFoundError(f"No videos found in {video_folder}")
    return mp4s[-1]

def show_videos(video_folder, width=640):
    for mp4 in sorted(glob.glob(os.path.join(video_folder, "*.mp4"))):
        data = open(mp4, "rb").read()
        b64 = base64.b64encode(data).decode("ascii")
        display(HTML(f"""
            <h4>{os.path.basename(mp4)}</h4>
            <video width={width} controls src="data:video/mp4;base64,{b64}"></video>
        """))

agent_files = {
    "sac_agent_simple_true_hard.pkl":  "SAC_true_obs",
    "sac_agent_normal_true_hard.pkl":  "SAC_MPSC_true_obs",
    "sac_agent_simple_hard.pkl":       "SAC_noisy_obs",
    "sac_agent_normal_hard.pkl":       "SAC_MPSC_noisy_obs",
    "sac_agent_complex_hard.pkl":      "SAC_MPSC_NF_noisy_obs"
}

# 4) Common “hard” roundabout config
cfg_hard = {
    "vehicles_count": 20,
    "observation": {
        "type":           "Kinematics",
        "vehicles_count": 8,
        "features":       ["x","y","vx","vy"],
    },
    "action":             {"type": "DiscreteMetaAction"},
    "duration":           40,
    "other_vehicles_type":"highway_env.vehicle.behavior.IDMVehicle",
}

for pickle_path, label in agent_files.items():
    # fresh env each time
    env = gym.make("roundabout-v0", render_mode="rgb_array", config=cfg_hard)
    with open(pickle_path, "rb") as f:
        agent = pickle.load(f)

    folder = f"videos/{label}"
    video_path = record_videos(env, folder, agent=agent, fps=15)
    print(f"Recorded: {video_path}")

show_videos("videos")
```
